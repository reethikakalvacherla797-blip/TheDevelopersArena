```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
df.info
```

```python
#Convert Date Column
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Line Plot (Sales Trend)
sns.lineplot(x='Date', y='Total_Sales', data=df)
plt.xticks(rotation=45)
plt.title('Sales Trend Over Time')
plt.show()
```

```python
#Bar Plot (Product Performance)
sns.barplot(x='Product', y='Total_Sales', data=df)
plt.title('Sales by Product')
plt.show()
```

```python
#Region Analysis
sns.barplot(x='Region', y='Total_Sales', data=df)
plt.title('Sales by Region')
plt.show()
```

```python
#Box Plot
sns.boxplot(x='Product', y='Total_Sales', data=df)
plt.title('Sales Distribution by Product')
plt.show()
```

```python
#Heatmap
sns.heatmap(df[['Quantity','Price','Total_Sales']].corr(),
            annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Plotly Interactive Chart
import plotly.express as px

fig = px.scatter(df,
                 x='Quantity',
                 y='Total_Sales',
                 color='Product',
                 hover_data=['Region'])
fig.show()
```
