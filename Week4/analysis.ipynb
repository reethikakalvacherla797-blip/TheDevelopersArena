```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv("sales_data.csv")
```

```python
data.head()
```

```python
# Data Cleaning
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
total_sales = data.groupby("Product")["Total_Sales"].sum()
print("\nTotal Sales by Product:")
print(total_sales)
```

```python
# Bar Chart
plt.figure(figsize=(6,4))
total_sales.plot(kind="bar")
plt.title("Sales by Product Category")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(6,6))
total_sales.plot(kind="pie", autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.ylabel("")
plt.show()
```

```python
# Convert Date column to datetime
data["Date"] = pd.to_datetime(data["Date"])
# Group total sales by date
sales_by_date = data.groupby("Date")["Total_Sales"].sum()
# Line Chart
plt.figure(figsize=(8,5))
plt.plot(sales_by_date.index, sales_by_date.values, marker="o")
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
